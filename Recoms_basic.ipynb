```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('book1-100k.csv')
```

```python
df.head(5)
```

```python
df_copy=df.copy()
```

```python
df_copy['RatingDist1']=df_copy['RatingDist1'].map(lambda x: x.lstrip('1:'))
```

```python
df_copy['RatingDist2']=df_copy['RatingDist2'].map(lambda x: x.lstrip('2:'))
df_copy['RatingDist3']=df_copy['RatingDist3'].map(lambda x: x.lstrip('3:'))
df_copy['RatingDist4']=df_copy['RatingDist4'].map(lambda x: x.lstrip('4:'))
df_copy['RatingDist5']=df_copy['RatingDist5'].map(lambda x: x.lstrip('5:'))
```

```python
df_copy.head(5)
```

```python
df_copy.drop(columns=['RatingDistTotal', 'Rating'])
```

```python
df_copy.dtypes
```

```python
df_copy['RatingDist1']=pd.to_numeric(df_copy['RatingDist1'])
df_copy['RatingDist2']=pd.to_numeric(df_copy['RatingDist2'])
df_copy['RatingDist3']=pd.to_numeric(df_copy['RatingDist3'])
df_copy['RatingDist4']=pd.to_numeric(df_copy['RatingDist4'])
df_copy['RatingDist5']=pd.to_numeric(df_copy['RatingDist5'])
```

```python
df_copy
```

```python
df_copy.drop(columns=['RatingDistTotal', 'Rating'])
```

```python
df_copy['ratingsTotal']=df_copy['RatingDist1']+df_copy['RatingDist2']+df_copy['RatingDist3']+df_copy['RatingDist4']+df_copy['RatingDist5']
```

```python
df_copy
```

```python
df_copy=df_copy.drop(columns=['RatingDistTotal', 'Rating'])
```

```python
df_copy
```

```python
df_copy.dtypes
```

```python
df_copy['Average_Rating']=(df_copy['RatingDist1']*1+df_copy['RatingDist2']*2+df_copy['RatingDist3']*3+df_copy['RatingDist4']*4+df_copy['RatingDist5']*5)/(df_copy['ratingsTotal'])
```

```python
df_copy
```

```python
df_copy.dtypes
```

# *Authors with most books*

```python
most_books=df_copy.groupby('Authors')['Name'].count().reset_index().sort_values('Name', ascending=False).head(10).set_index('Authors')
```

```python
most_books.head(10)
```

```python
sns.set_context('talk')
plt.figure(figsize=(30,10))
res=sns.barplot(most_books.index, most_books['Name'])
res.set_title('Top 10 authors with most books')
res.set_ylabel('Total Number Of Books')
```

# ***Highest Rated Authors***
